```python
df_anthro_orig = pd.read_sql_table('ANTHRO_FEAT', con=engine)
df_physical_orig = pd.read_sql_table('PHYSICAL_FEAT', con=engine)
df_shots_orig = pd.read_sql_table('SHOTS', con=engine)
df_events_orig = pd.read_sql_table('EVENTS', con=engine)
```

```python
df_events = df_events_orig.copy()
df_events.columns = ['GAME_ID', 'EVENTNUM', 'EVENTMSGTYPE', 'EVENTMSGACTIONTYPE', 'PERIOD',
                     'WCTIMESTRING', 'PCTIMESTRING', 'HOMEDESCRIPTION', 'NEUTRALDESCRIPTION',
                     'VISITORDESCRIPTION', 'SCORE', 'SCOREMARGIN', 'PERSON1TYPE',
                     'PLAYER1_ID', 'PLAYER1_NAME', 'PLAYER1_TEAM_ID', 'PLAYER1_TEAM_CITY',
                     'PLAYER1_TEAM_NICKNAME', 'PLAYER1_TEAM_ABBREVIATION', 'PERSON2TYPE',
                     'PLAYER2_ID', 'PLAYER2_NAME', 'PLAYER2_TEAM_ID', 'PLAYER2_TEAM_CITY',
                     'PLAYER2_TEAM_NICKNAME', 'PLAYER2_TEAM_ABBREVIATION', 'PERSON3TYPE',
                     'PLAYER3_ID', 'PLAYER3_NAME', 'PLAYER3_TEAM_ID', 'PLAYER3_TEAM_CITY',
                     'PLAYER3_TEAM_NICKNAME', 'PLAYER3_TEAM_ABBREVIATION']
df_events
```

```python
df_shots = df_shots_orig.copy()
df_shots["ACTION_TYPE_CAT"] = df_shots["ACTION_TYPE"].astype('category').cat.codes
df_shots["MADE_SHOT"] = df_shots["SHOT_MADE_FLAG"].astype('bool')
df_shots["IS_3PT"] = df_shots["SHOT_TYPE"].str.contains("3PT")
df_shots["HTM"] = df_shots["HTM"].astype('category')
df_shots["VTM"] = df_shots["VTM"].astype('category')

cols_to_drop = ["EVENT_TYPE", "GRID_TYPE", "PERIOD", "SHOT_ATTEMPTED_FLAG", "SHOT_MADE_FLAG", "SHOT_TYPE",
                "SHOT_ZONE_AREA", "SHOT_ZONE_BASIC", "SHOT_ZONE_RANGE", "TEAM_NAME", "TEAM_ID"]

df_shots = df_shots.drop(columns=cols_to_drop)

df_shots
```
